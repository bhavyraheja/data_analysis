```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the CSV file
df = pd.read_csv("Global-Superstore - Global-Superstore.csv.csv")

print(df.head())
```

```python
# Example 1: Line Plot
plt.figure(figsize=(10, 6))
plt.plot(df['Sales'], df['Shipping Cost'])
plt.title('Line Plot of Sales vs Shipping Cost')
plt.xlabel('Sales')
plt.ylabel('Shipping Cost')
plt.grid(True)
plt.show()
```

```python
# Example 2: Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Sales'], df['Shipping Cost'])
plt.title('Scatter Plot of Sales vs Shipping Cost')
plt.xlabel('Sales')
plt.ylabel('Shipping Cost')
plt.grid(True)
plt.show()
```

```python
# Example 3: Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Sales'], bins=30, edgecolor='black')
plt.title('Histogram of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Example 4: Seaborn Pair Plot (useful for multiple columns)
sns.pairplot(df[['Sales', 'Shipping Cost', 'Profit']])
plt.show()
```
