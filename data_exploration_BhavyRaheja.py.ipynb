```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Global-Superstore - Global-Superstore.csv.csv")

# Print the first few rows of the DataFrame
print(df.head(5))
```

```python
df.tail()
```

```python
#Info: View a concise summary of the data
df.info()
```

```python
#Describe:View summary statistics for the data
df.describe()
```

```python
#Data Cleaning

#Drop duplicates: Remove duplicate rows
df.drop_duplicates(inplace=True)
#Handle missing values: Replace missing values with a specific value (e.g. mean, median, imputed value)
df.fillna(df.mean(), inplace=True)
```

```python
#Data Visualization

#Histogram: Plot a histogram for a specific column
sns.histplot(df['Sales'])
plt.show()
```

```python
#Scatter plot: Plot a scatter plot for two columns
sns.scatterplot(x='Row ID', y='Postal Code', data=df)
plt.show()
```

```python
#Correlation matrix: Plot a correlation matrix for the data
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', square=True)
plt.show()
```

```python
#Data Transformation

#Group by: Group the data by a specific column and perform an aggregation operation
df.groupby('Sales').mean()
```

```python
#Pivot table: Create a pivot table for the data
pd.pivot_table(df, values='Sales', index='Row ID', columns='Postal Code')
```
